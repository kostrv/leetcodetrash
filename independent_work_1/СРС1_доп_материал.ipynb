```python
import time
import matplotlib.pyplot as plt

# функция для теста
def simple_sum(n):
    s = 0
    for i in range(n):
        s += i
    return s


sizes = [1000, 5000, 10000, 50000]
times = []

for n in sizes:

    start = time.time()

    simple_sum(n)

    end = time.time()

    execution_time = end - start

    times.append(execution_time)


# построение графика
plt.plot(sizes, times, marker='o')

plt.xlabel("Размер входных данных (n)")
plt.ylabel("Время выполнения (секунды)")
plt.title("Пример измерения времени выполнения функции")

plt.grid(True)

plt.show()
```

```python
import time
import matplotlib.pyplot as plt

# алгоритм 1 (пример O(n))
def linear_function(n):
    s = 0
    for i in range(n):
        s += i
    return s


# алгоритм 2 (пример O(n²))
def quadratic_function(n):
    s = 0
    for i in range(n):
        for j in range(n):
            s += 1
    return s


sizes = [1000, 5000, 10000, 20000]

times_linear = []
times_quadratic = []


for n in sizes:

    # измерение для первого алгоритма
    start = time.time()
    linear_function(n)
    end = time.time()
    times_linear.append(end - start)

    # измерение для второго алгоритма
    start = time.time()
    quadratic_function(n)
    end = time.time()
    times_quadratic.append(end - start)


# построение графика
plt.plot(sizes, times_linear, marker='o', label='Linear function O(n)')
plt.plot(sizes, times_quadratic, marker='o', label='Quadratic function O(n²)')

plt.xlabel("Размер входных данных (n)")
plt.ylabel("Время выполнения (секунды)")
plt.title("Сравнение времени выполнения функций")

plt.legend()
plt.grid(True)

plt.show()
```
