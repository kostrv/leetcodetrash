```python
from collections import Counter
import heapq
```

```python
# создаю класс для узлов дерева Хаффмана, который будет использоваться для построения дерева и кодирования символов
class Node:
    def __init__(self, char, freq):
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None

    def __lt__(self, other): # для сравнения узлов в приоритетной очереди использую магический метод __lt__
        return self.freq < other.freq
```

```python

def huffman_encode(text : str):
    if not text: return '', None

    # cчитаю частоты символов
    frequencies : dict = Counter(text)
    
    # создаю очередь FIFO для узлов
    heap : list = [Node(char, freq) for char, freq in frequencies.items()]
    heapq.heapify(heap)
    
    while len(heap) > 1:
        left = heapq.heappop(heap) # извлекаю два узла с наименьшей частотой
        right = heapq.heappop(heap)
        
        # создаю новый внутренний узел, который объединяет эти два узла, и добавляю его обратно в очередь
        merged = Node(char=None, freq= left.freq + right.freq) 
        merged.left = left
        merged.right = right
        
        heapq.heappush(heap, merged)
    
    root = heap[0]
    
    # создаю внутреннюю функцию для генерации таблицы кодов (рекурсивный обход дерева)
    codes = {}
    def generate_codes(node : Node, current_code : str):
        if node is None: return
        
        if node.char is not None:
            codes[node.char] = current_code
            return
        
        generate_codes(node.left, current_code + '0')
        generate_codes(node.right, current_code + '1')
    
    generate_codes(root, '')
    
    # кодировка текста, заменяя каждый символ на его код из таблицы
    encoded_text = ''.join(codes[char] for char in text)
    
    return encoded_text, root
```

```python

message = 'codetestcccode12343'
encoded_str, tree_root = huffman_encode(message)

print(f'исходное сообщение: {message}')
print(f'закодированное: {encoded_str}')
```

```python
def huffman_decode(encoded_text : str, root : Node):
    # опорный элемент root обязательно должен быть передан, иначе декодировать не по чему
    if not encoded_text or not root:
        return ''
    
    decoded_text = []
    current_node = root
    
    # перебираю каждый бит в закодированной строке
    for bit in encoded_text:
        # Если бит '0' - в левое поддерево
        # Если бит '1' - в правое поддерево
        if bit == '0':
            current_node = current_node.left
        else:
            current_node = current_node.right
            
        # проверяю, содержит ли символм текущий узел
        if current_node.char is not None:
            # симол найден - добавляю его в результат
            decoded_text.append(current_node.char)
            # возвращаюсь в корень дерева для поиска следующего символа
            current_node = root
            
    return ''.join(decoded_text)
```

```python
print(f'исходное сообщение: {message}')
print(f'закодированное: {encoded_str}')
print(f'декодированное: {huffman_decode(encoded_str, tree_root)}')
```
